```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/mihirshah/Desktop/DATASETS/Generation_Data.csv')
df.drop(['PLANT_ID'], axis = 1, inplace = True)
df.drop(['DATE_TIME'], axis = 1, inplace = True)


DC_POWER = df['DC_POWER']
AC_POWER = df['AC_POWER']
DAILY_YIELD = df['DAILY_YIELD']
TOTAL_YIELD = df['TOTAL_YIELD']

df['SOURCE_KEY'] = df['SOURCE_KEY'].astype('object') 
df['DC_POWER'] = df['DC_POWER'].astype('int64') 
df['AC_POWER'] = df['AC_POWER'].astype('int64') 
df['DAILY_YIELD'] = df['DAILY_YIELD'].astype('int64') 
df['TOTAL_YIELD'] = df['TOTAL_YIELD'].astype('int64')   
df.drop(['DC_POWER', 'AC_POWER' ,'DAILY_YIELD', 'TOTAL_YIELD'], axis = 1, inplace = True)

df.head(5)
```

```python
dataset = pd.get_dummies(df)
```

```python
dataset['DC_POWER'] = DC_POWER
dataset['AC_POWER'] = AC_POWER
dataset['DAILY_YIELD'] = DAILY_YIELD
dataset['TOTAL_YIELD'] = TOTAL_YIELD


dataset['DC_POWER'] = dataset['DC_POWER'].astype('int64') 
dataset['AC_POWER'] = dataset['AC_POWER'].astype('int64') 
dataset['DAILY_YIELD'] = dataset['DAILY_YIELD'].astype('int64') 
dataset['TOTAL_YIELD'] = dataset['TOTAL_YIELD'].astype('int64')  
dataset.to_csv('/Users/mihirshah/Desktop/DATASETS/Generation_Data(123).csv')
dataset
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
y = y.reshape(len(y), 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

## Training the Regression model

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred = y_pred.astype('int64')
y_test = y_test.astype('int64')
y_pred1 = y_pred.reshape(len(y_pred), 1)
y_test1 = y_test.reshape(len(y_test), 1)
a = np.concatenate((y_pred1, y_test1), 1)
am = pd.DataFrame(data  = a, columns = ('Predicted Value', 'Real Value'))
am.to_csv('/Users/mihirshah/Desktop/DATASETS/Generation_Data(123).csv')
```

#### r^2 Metrics to evaluate models performance

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
